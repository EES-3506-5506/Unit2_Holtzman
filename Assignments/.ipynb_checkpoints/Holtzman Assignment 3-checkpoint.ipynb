11. Write a function to convert from miles to kilometers (km) and km to miles, with the
user specifying either miles or km as inputs.

```python
def conversion (distance, to_units='miles'):  #this is a function that turns km to miles and miles to km
    if to_units == 'miles' :                  #conditional statements-one for converting TO miles and one for TO km
        conv_distance = distance*0.621
    if to_units == 'km' :
        conv_distance = distance*1.609

    return conv_distance      #whichever statement is applicable based on the information provided/entered runs and we return the output specified
```

14. Write a function that computes and returns the surface area of a sphere.

```python
def sphere_sa (d): # this function takes a diameter measurement of a sphere and returns the surface area
    surface_area= 4 * 3.14 * ((d/2)**2)   #math to calculate surface area
    return surface_area
```

16. Write a function that takes two arguments (feet and inches) and returns height in
meters.

```python
def height(feet, inches):            #function to calculate someones height in meters given their height in feet and inches
    total_inches= feet* 12 + inches  #calculate the total amount of inches based on given parameters
    meters= total_inches*0.0254      #calculate meters from total inches
    return meters
```

```python
height (5, 7)
```

17. Write a function that takes that takes one argument (height in meters) and returns
two arguments (feet and inches).

```python
def us_height (meters):        #function that converts meters into feet and inches
    inch=meters*39.3701        #math to turn meters to feet
    feet=inch/12               #math to turn feet to inches
    return (feet, inch)
```
